# Auxiliary functions

```python
def invest(day, day20, day21, day22, total, total_asset, position_dict, top_stocks):

    if day != 0:
        uninvested = 0
    elif (day == 0) & (first_run == False):
        uninvested = 0
    elif (day == 0) & (first_run == True):
        uninvested = 1
    to_hold = []
    to_delete = []

    for key in position_dict:
        # Sell
        if key not in top_stocks:
            uninvested += position_dict[key]
            to_delete.append(key)
#             print(f'Sell {key} on {day21} of market open')
        # Hold
        else:
            to_hold.append(key)

    # Remove sold stocks from position
    for key in to_delete:
        del position_dict[key]

#     print(f'Hold {to_hold} on {day21} market open')

    # Buy stocks
    for index in list(set(top_stocks) - set(to_hold)):
        position_dict[index] = uninvested / (5 - len(to_hold))
#         print(f'Buy {index} on {day21} market open')

    # Calculate return right away
    for key in top_stocks:
        percent_change = x_test[(x_test['datadate'] == day20) & (x_test['tic'] == key)]['ret_d'].values[0]
        total += position_dict[key] * percent_change
        # This calculates the position on the next day
        position_dict[key] = position_dict[key] * (1 + percent_change)
#     print(f'position on {day22} market open will be {position_dict}')

    total_asset.append(total)
    print(f'Total asset on {day22} will be {total}')
#     print(f'It should be the same as {sum(position_dict.values())}')
    
    return total_asset, total, position_dict
```

```python
# Loss with scce
scce = tf.keras.losses.SparseCategoricalCrossentropy(reduction='none')
def custom_loss(y_true, y_pred, ret_d):
    y_true = tf.cast(y_true, dtype=tf.float32)
    loss = tf.reduce_mean(scce(y_true, y_pred) * tf.abs(ret_d))
    return loss

# Numpy version of loss
def custom_loss_np(y_true, y_pred, ret_d):
    loss = np.mean(scce(y_true, y_pred) * abs(ret_d))
    return loss
```

## OLS once

```python
# Set OLS factors
OLS_dict = {}
OLS_factors_list = [['Mom_3day', 'mkt_cap'], ['Mom_5day', 'mkt_cap'], ['Mom_3day', 'mkt_cap', 'open/MA10'], 
                    ['Mom_3day', 'mkt_cap', 'RSI'], ['Mom_3day', 'RSI', 'MACD'], ['Mom_3day', 'RSI']]

# OLS_factors_list = [['Mom_3day', 'mkt_cap'], ['Mom_5day', 'mkt_cap'], ['Mom_3day', 'mkt_cap', 'open/MA10'], 
#                     ['Mom_3day', 'mkt_cap', 'RSI'], ['Mom_3day', 'mkt_cap', 'MACD'], ['Mom_3day', 'RSI', 'MACD'],
#                     ['Mom_3day', 'RSI'], ['Mom_3day', 'MACD']]
```

```python
# Initialize portfolio_dict
portfolio_dict = {}
for i in range(len(OLS_factors_list)):
    portfolio_dict[i] = pd.DataFrame()

# Initialize SR-related variables
SR_dict = {}
for i in range(len(OLS_factors_list)):
    SR_dict[i] = pd.DataFrame(columns=['pred_score', 'avg_excess_ret', 'std_excess_ret', 'SR'])

pred_avg_dict = {}
excess_return_dict = {}
for i in range(len(OLS_factors_list)):
    pred_avg_dict[i] = 0
    excess_return_dict[i] = np.zeros((1, seq_length * num_iters))
```

```python
all_OLS_factors = []
for OLS_factors in OLS_factors_list:
    for OLS_factor in OLS_factors:
        if OLS_factor not in all_OLS_factors:
            all_OLS_factors.append(OLS_factor)
print(all_OLS_factors)

data = data[all_OLS_factors + ['datadate', 'tic', 'conm', 'ret_d', 'excess_ret_d']]
print(data.columns)
data
```

```python
ftd = 0
ltd = 99
seq_length = 20
graph = True
num_stocks = 5
first_run = True
total_dict = {}
total_asset_dict = {}
position_dict_all = {}
for i in range(len(OLS_factors_list)):
    total_dict[i] = 1
    total_asset_dict[i] = [1]
    position_dict_all[i] = {}

data_train = data[(data['datadate'] >= all_days[ftd]) & (data['datadate'] <= all_days[ltd])]
print(f'Training from {all_days[ftd]} to {all_days[ltd]} and has shape {data_train.shape}')

ftestd = ltd + 1
ltestd = ltd + 20
data_test = data[(data['datadate'] >= all_days[ftestd]) & (data['datadate'] <= all_days[ltestd])]
print(f'Testing from {all_days[ftestd]} to {all_days[ltestd]} and has shape {data_test.shape}')

y_train = np.array(data_train['ret_d'])
y_test = np.array(data_test['ret_d'])

for i in range(len(OLS_factors_list)):
    
    OLS_factors = OLS_factors_list[i]
    x_train = np.array(data_train[OLS_factors])
    x_test = data_test[OLS_factors + ['datadate', 'tic', 'ret_d']]
    OLS_dict[i] = LinearRegression().fit(x_train, y_train)
    
    pred_avg = 0
    excess_return = np.zeros((1, seq_length))
    
    for day in range(seq_length):
        test_day = all_days[ftestd + day]
        buy_day = all_days[ftestd + day + 1]
        simul_day = all_days[ftestd + day + 2]
        x_test_temp = np.array(x_test[x_test['datadate'] == test_day][OLS_factors])
        y_pred = OLS_dict[i].predict(x_test_temp)
        top_indices = np.argsort(y_pred)[-num_stocks:]
        top_stocks = [num_to_tic_dict[num] for num in top_indices]
        print(f'top_stocks by OLS factors {OLS_factors} to buy on {buy_day} are {top_stocks}')
        
        # Simulate investment
        total_asset, total, position_dict = invest(day, test_day, buy_day, simul_day, total_dict[i],
                                                   total_asset_dict[i], position_dict_all[i], top_stocks)
        total_dict[i] = total
        total_asset_dict[i] = total_asset
        position_dict_all[i] = position_dict   
        
        # Calculate SR PT1
        pred = sum([y_pred[ind] for ind in top_indices]) / num_stocks
        pred_avg_dict[i] += pred
        pred_avg += pred
        excess_ret_d = data_test[(data_test['datadate'] == test_day) & (data_test['tic'].isin(top_stocks))]['excess_ret_d'].values.mean()
        excess_return[0, day] = excess_ret_d
        excess_return_dict[i][0, day] = excess_ret_d
        
    # Calculate SR PT2
    pred_avg /= seq_length
    excess_return_avg = np.mean(excess_return)
    excess_return_std = np.std(excess_return)

    key = f'{all_days[ltd+1]}-{all_days[ltd+seq_length]}'
    SR_dict[i].loc[key, 'pred_score'] = pred_avg
    SR_dict[i].loc[key, 'avg_excess_ret'] = excess_return_avg
    SR_dict[i].loc[key, 'std_excess_ret'] = excess_return_std
    SR_dict[i].loc[key, 'SR'] = excess_return_avg / excess_return_std * np.sqrt(252)
    print(f"OLS {i} has SR {SR_dict[i].loc[key, 'SR']}")
    
for i in range(len(OLS_factors_list)):
    plt.plot(all_days[ftestd-1:ltestd+1], total_asset_dict[i], label=f'OLS_{i}')

plt.xticks(rotation=45)
plt.xticks(ticks=range(0, 22, 4), labels=[all_days[j] for j in range(ftestd-1, ltestd+2, 4)])
plt.legend()
plt.show()
```

## Refit OLS every `seq_length` days

```python
ftd = 0
ltd = 99
seq_length = 20
num_stocks = 5
first_run = False

for num_iter in range(1, num_iters):
    
    print(f'Running {num_iter + 1} of {num_iters} iterations. First iteration already finished')
    
    ftd += 20 
    ltd += 20
    
    data_train = data[(data['datadate'] >= all_days[ftd]) & (data['datadate'] <= all_days[ltd])]
#     print(f'Training from {all_days[ftd]} to {all_days[ltd]} and has shape {data_train.shape}')

    ftestd = ltd + 1
    ltestd = ltd + 20
    data_test = data[(data['datadate'] >= all_days[ftestd]) & (data['datadate'] <= all_days[ltestd])]
#     print(f'Testing from {all_days[ftestd]} to {all_days[ltestd]} and has shape {data_test.shape}')

    y_train = np.array(data_train['ret_d'])
    y_test = np.array(data_test['ret_d'])

    for i in range(len(OLS_factors_list)):

        OLS_factors = OLS_factors_list[i]
        x_train = np.array(data_train[OLS_factors])
        x_test = data_test[OLS_factors + ['datadate', 'tic', 'ret_d']]
        OLS_dict[i] = LinearRegression().fit(x_train, y_train)

        pred_avg = 0
        excess_return = np.zeros((1, seq_length))

        for day in range(seq_length):
            test_day = all_days[ftestd + day]
            buy_day = all_days[ftestd + day + 1]
            simul_day = all_days[ftestd + day + 2]
            x_test_temp = np.array(x_test[x_test['datadate'] == test_day][OLS_factors])
            y_pred = OLS_dict[i].predict(x_test_temp)
            top_indices = np.argsort(y_pred)[-num_stocks:]
            top_stocks = [num_to_tic_dict[num] for num in top_indices]
#             print(f'top_stocks by OLS factors {OLS_factors} to buy on {buy_day} are {top_stocks}')

            # Simulate investment
            total_asset, total, position_dict = invest(day, test_day, buy_day, simul_day, total_dict[i],
                                                       total_asset_dict[i], position_dict_all[i], top_stocks)
            total_dict[i] = total
            total_asset_dict[i] = total_asset
            position_dict_all[i] = position_dict   

            # Calculate SR PT1
            pred = sum([y_pred[ind] for ind in top_indices]) / num_stocks
            pred_avg_dict[i] += pred
            pred_avg += pred
            excess_ret_d = data_test[(data_test['datadate'] == test_day) & (data_test['tic'].isin(top_stocks))]['excess_ret_d'].values.mean()
            excess_return[0, day] = excess_ret_d
            excess_return_dict[i][0, num_iter * seq_length + day] = excess_ret_d
            
        # Calculate SR PT2
        pred_avg /= seq_length
        excess_return_avg = np.mean(excess_return)
        excess_return_std = np.std(excess_return)

        key = f'{all_days[ltd+1]}-{all_days[ltd+seq_length]}'
        SR_dict[i].loc[key, 'pred_score'] = pred_avg
        SR_dict[i].loc[key, 'avg_excess_ret'] = excess_return_avg
        SR_dict[i].loc[key, 'std_excess_ret'] = excess_return_std
        SR_dict[i].loc[key, 'SR'] = excess_return_avg / excess_return_std * np.sqrt(252)
        print(f"OLS {i} has SR {SR_dict[i].loc[key, 'SR']}")

    if num_iter % 10 == 0 or num_iter == num_iters - 1:
        plt.figure(figsize=(16, 6))
        for i in range(len(OLS_factors_list)):
            x_axis = all_days[ltestd+1-len(total_asset_dict[i]):ltestd+1]
            plt.plot(x_axis, [math.log(val, 2) for val in total_asset_dict[i]], label=f'OLS_{i}')

        indices_to_display = np.linspace(0, len(x_axis)-1, 15, dtype=int)
        plt.xticks(indices_to_display, [x_axis[i] for i in indices_to_display], rotation=45)
        plt.title(f'Log Plot of Total Asset Simulated by OLS ({num_stocks} stocks per day)')
        plt.ylabel('Log of Total Asset')
        plt.xlabel('Date')
        plt.grid()
        plt.legend()
        plt.show()
```

```python
# Calculate the overall SR
SR_whole_period = pd.DataFrame(columns=['pred', 'avg_excess_ret', 'std_excess_ret', 'SR'])

for i in range(len(OLS_factors_list)):
    pred_avg_dict[i] /= num_iters * seq_length
    excess_return_avg = np.mean(excess_return_dict[i][0])
    excess_return_std = np.std(excess_return_dict[i][0])
    
    # Calculate the SR table for original models
    SR_whole_period.loc[f'OLS_{i}', 'pred'] = pred_avg_dict[i]
    SR_whole_period.loc[f'OLS_{i}', 'avg_excess_ret'] = excess_return_avg
    SR_whole_period.loc[f'OLS_{i}', 'std_excess_ret'] = excess_return_std
    SR_whole_period.loc[f'OLS_{i}', 'SR'] = excess_return_avg / excess_return_std * np.sqrt(252)
print(SR_whole_period)
```
